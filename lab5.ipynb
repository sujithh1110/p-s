```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
```

```python
np.random.seed(42)
N = 1000  # Number of households
data = np.random.choice([0, 1, 2, 3, 4, 5], size=N, p=[0.1, 0.2, 0.3, 0.2, 0.15, 0.05])
```

```python
values, counts = np.unique(data, return_counts=True)
pmf = counts / N
```

```python
cdf = np.cumsum(pmf)

```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(values, pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Vehicles")
plt.ylabel("Probability")
plt.title("PMF of Vehicle Ownership")
```

```python
plt.subplot(1, 2, 2)
plt.plot(values, cdf, marker='o', linestyle='-', color='red')
plt.xlabel("Number of Vehicles")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Vehicle Ownership")

plt.tight_layout()
plt.show()
```

```python
N = 100
p = 0.05


prob_10 = binom.pmf(10, N, p)
print(f"P(Exactly 10 defective gadgets) = {prob_10:.4f}")


prob_15_or_more = 1 - binom.cdf(14, N, p)
print(f"P(At least 15 defective gadgets) = {prob_15_or_more:.4f}")


expected_defective = N * p
print(f"Expected number of defective gadgets: {expected_defective}")
```
