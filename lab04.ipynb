```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
  df=pd.read_csv(file_path)
  return df
```

```python
def plot_histogram(df,column,bins=10):
  plt.figure(figsize=(8,6))
  sns.histplot(df[column],bins=bins,kde=True,color='blue')
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

```python
def plot_scatter(df,x_column,y_column):
  plt.figure(figsize=(8,6))
  sns.scatterplot(x=df[x_column],y=df[y_column],color='red')
  plt.title(f'Scatter Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
def plot_bar_chart(df,x_column,y_column):
  plt.figure(figsize=(10,6))
  sns.barplot(x=df[x_column],y=df[y_column],palette='viridis')
  plt.title(f'Bar Chart of {y_column} by {x_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.xticks(rotation=45)
  plt.show()
```

```python
#load the dataset
file_path='titanic_dataset_250-7.csv'
data=load_data(file_path)
```

```python
plot_histogram(data,'Age',bins=20)
```

```python
plot_scatter(data,'Age','Fare')
```

```python
plot_scatter(data,'Age','Survived')
```
