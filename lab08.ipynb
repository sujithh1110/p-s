```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Dataset
study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])

```

```python
# Function to plot scatter plot
def plot_scatter(x, y, xlabel, ylabel, title):
    plt.scatter(x, y, color='b', alpha=0.7)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()

# Scatter Plots
plot_scatter(study_hours, exam_scores, "Study Hours", "Exam Scores", "Study Hours vs Exam Scores")
plot_scatter(sleep_hours, exam_scores, "Sleep Hours", "Exam Scores", "Sleep Hours vs Exam Scores")
plot_scatter(study_hours, sleep_hours, "Study Hours", "Sleep Hours", "Study Hours vs Sleep Hours")

# Function to calculate Pearson correlation coefficient
def calculate_pearson(x, y, label_x, label_y):
    correlation, _ = pearsonr(x, y)
    print(f"Pearson Correlation ({label_x} vs {label_y}): {correlation:.4f}")

# Calculate Pearson Correlation
calculate_pearson(study_hours, exam_scores, "Study Hours", "Exam Scores")
calculate_pearson(sleep_hours, exam_scores, "Sleep Hours", "Exam Scores")
calculate_pearson(study_hours, sleep_hours, "Study Hours", "Sleep Hours")
```
